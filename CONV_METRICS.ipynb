```python
# Extraccion de los valores de la metrica Talk_TIme ( Metric = 23)
Kpi_Metric = data[data["METRIC"].notna()]
Kpi_Talk_Time = Kpi_Metric[Kpi_Metric["METRIC"] == 23]
Kpi_Talk_TIme = Kpi_Talk_Time["METRIC"].value_counts()
```

```python
# Extraccion del numero total de registros de la  metrica Talk_TIme ( Metric = 23)
Kpi_Count = Kpi_Talk_TIme
```

```python
print(Kpi_Count)
```

```python
# Extraccion del valor total de la metrica Talk_TIme ( Metric = 23)
Kpi_Value = Kpi_Talk_Time["VALUE"].sum()
```

```python
print(Kpi_Value)
```

```python
# función ara sacar KPI de Talk_TIme
def Kpi_Talk_TIme(Kpi_Value, Kpi_Count):
  return Kpi_Value / Kpi_Count
```

```python
Kpi_Talk_TIme(Kpi_Value, Kpi_Count)
```
